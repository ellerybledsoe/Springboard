```python
# binning numeric scores by letter grade
# move this to appropriate directory for storing functions and read the function in from there

def letter_grade(row):
    if row['Score'] >= 90:
        return 'A'
    if row['Score'] >= 80:
        return 'B'
    if row['Score'] >= 70:
        return 'C'
    return 'F'
```

```python
# move this to functions directory and call it from another functions script

def dummies(df):
    dummies = pd.DataFrame()
    for col in df.columns.tolist():
        dummies = pd.concat([dummies, pd.get_dummies(df[col], dummy_na=True, prefix=col)], axis = 1)
    return dummies
```

```python
# function to do multiple aggregations that will flatten the data

def flatten(x):
    d = {}
    d['reviews_count'] = x['review_id'].count()
    d['useful_sum'] = x['useful'].sum()
    d['funny_sum'] = x['funny'].sum()
    d['cool_sum'] = x['cool'].sum()
    d['stars_1.0_sum'] = x['stars_1.0'].sum()
    d['stars_2.0_sum'] = x['stars_2.0'].sum()
    d['stars_3.0_sum'] = x['stars_3.0'].sum()
    d['stars_4.0_sum'] = x['stars_4.0'].sum()
    d['stars_5.0_sum'] = x['stars_5.0'].sum()
    d['stars_nan_sum'] = x['stars_nan'].sum()
    return pd.Series(d, index=['reviews_count', 'useful_sum', 'funny_sum', 'cool_sum','stars_1.0_sum','stars_2.0_sum','stars_3.0_sum','stars_4.0_sum','stars_5.0_sum','stars_nan_sum'])


# https://stackoverflow.com/questions/14529838/apply-multiple-functions-to-multiple-groupby-columns/53096340
```

```python
# flattening yelp reviews data 
# ** takes a few minutes to run **

print(yelp_reviews_tx1.business_id.nunique())
yelp_reviews_tx2 = yelp_reviews_tx1.groupby(['business_id', 'date_trim']).apply(flatten)
yelp_reviews_tx2 = yelp_reviews_tx2.reset_index()
print(yelp_reviews_tx2.shape)
yelp_reviews_tx2.head()
```

```python
yelp_reviews_tx3 = yelp_reviews_tx2.reset_index()
#del yelp_reviews_tx3['index']
yelp_reviews_tx3.business_id.nunique()
```

```python
# converting Inspection Date column to datetime64 to support appropriate sorting for loop
from datetime import datetime as dt

austin_inspections_yelp_merge4 = austin_inspections_yelp_merge3_1.copy()
austin_inspections_yelp_merge4['Inspection Date'] = pd.to_datetime(austin_inspections_yelp_merge4['Inspection Date'])
austin_inspections_yelp_merge4.head()
```

```python
from datetime import datetime as dt

count = 0
#prior_date = dt.strptime('2017-10-17', "%Y-%m-%d").date()  # inspections to be roughly quarterly, thus allowing a 3-month lag for first set of yelp reviews
prior_date = dt.date(austin_inspections_yelp_merge4['Inspection Date'].min() - np.timedelta64(60,'D'))
loop_out = pd.DataFrame()

for date in pd.Series(austin_inspections_yelp_merge4['Inspection Date'].unique()).sort_values():
    #print('prior date: ',prior_date)
    #print('date: ',dt.date(date))
    #print(date)
    loop_inspections = austin_inspections_yelp_merge4.loc[austin_inspections_yelp_merge4['Inspection Date'].astype(str)==str(dt.date(date))] # filtering down to specific inspection date
    #print(loop_inspections.shape)
    loop_reviews = yelp_reviews_tx3.loc[(yelp_reviews_tx3['date_trim'] <= dt.date(date)) & (yelp_reviews_tx3['date_trim'] > prior_date)] # filtering reviews down to those that precede the inspection date
    #print(loop_reviews.shape)
    del loop_reviews['date_trim'] # dropping date column from reviews in order to rollup counts
    loop_reviews = loop_reviews.groupby(by=['business_id']).sum().reset_index() # rolling up review counts
    loop_inspections_reviews_merge = loop_inspections.merge(loop_reviews,how='left', on='business_id') # joining inspections and preceding review totals
    loop_out = pd.concat([loop_out,loop_inspections_reviews_merge])
    #print(loop_out.shape)
    if count%100 == 0:
        print(loop_out.shape)
    
    prior_date = dt.date(date - np.timedelta64(60,'D'))
    count+=1

inspections_reviews_merge = pd.DataFrame(loop_out)
print(inspections_reviews_merge.shape)
inspections_reviews_merge.head()
```
