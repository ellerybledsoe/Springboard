```python
# reading in Austin restaurant inspections data

austin_inspections = pd.read_csv('/Users/EBled1/Springboard DS/Springboard/Capstone Two/Capstone Two/data/raw/austin_restaurant_inspections.csv')
```

```python
# binning numeric scores by letter grade
# move this to appropriate directory for storing functions and read the function in from there

def letter_grade(row):
    if row['Score'] >= 90:
        return 'A'
    if row['Score'] >= 80:
        return 'B'
    if row['Score'] >= 70:
        return 'C'
    return 'F'
```

```python
austin_inspections['letter_grade'] = austin_inspections.apply (lambda row: letter_grade(row), axis=1)
# laustin_inspections.sort_values(by=['Restaurant Name']).head(10)
```

```python
# looking at distribution of records by letter grade

austin_inspections.letter_grade.value_counts(normalize=True)
```

```python
# extracting a cleaner address line item for merge

austin_inspections['addr_line_1'] = (austin_inspections['Address'].str.split('\n').str[0].str.upper())
austin_inspections.head(1)
```

```python
# converting 'Zip Code' from float to string for merge

austin_inspections['Zip Code'] = (austin_inspections['Zip Code'].fillna(0).apply(np.int64)).apply(str)
```

```python
austin_inspections.head(1)
```

```python
austin_inspections['Restaurant Name'] = austin_inspections['Restaurant Name'].str.upper()
yelp_tx['name'] = yelp_tx['name'].str.upper()
```

```python
#join inspections and yelp data

print(austin_inspections.shape)
austin_inspections_yelp = austin_inspections.merge(yelp_tx, how='left', left_on =['addr_line_1'], right_on =['address'],
                        suffixes = ('','_yelp'))
print(austin_inspections_yelp.shape)

# increased row count from join is addressed in subsequent filtering 
```
